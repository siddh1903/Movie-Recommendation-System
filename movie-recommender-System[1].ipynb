```python
# Define a conversion function to extract genre names
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Rebuild the similarity matrix
similarity = cosine_similarity(vectors)

# Verify the shape of the similarity matrix
print(f"Similarity Matrix Shape: {similarity.shape}")
```

```python
# Improved recommendation function with debug statements
def recommend(movie):
    # Normalize movie title by converting to lowercase
    movie_lower = movie.lower()
    
    # Match the movie title exactly, case insensitive
    matched_movies = new_df[new_df['title'].str.lower() == movie_lower]
    
    # Check if movie exists in the dataset
    if matched_movies.empty:
        print(f"Movie '{movie}' not found in the dataset!")
        return
    
    # Get the first movie index and print it for debugging
    movie_index = matched_movies.index[0]
    print(f"Movie Index for '{movie}': {movie_index}")
    
    # Get similarity scores for the movie
    distances = similarity[movie_index]
    
    # Print similarity scores for debugging
    print(f"Similarity Scores for '{new_df.iloc[movie_index]['title']}': {distances[:10]}")
    
    # Sort the movies based on similarity scores, excluding the queried movie itself
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Display the top 5 recommendations and print their indices for debugging
    print(f"Movies similar to '{new_df.iloc[movie_index]['title']}':")
    for i in movie_list:
        print(f"Index: {i[0]}, Movie: {new_df.iloc[i[0]]['title']}")

# Rebuild the similarity matrix if necessary
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

# Example: Recommend movies similar to 'Avatar'
recommend('Avatar')
```
